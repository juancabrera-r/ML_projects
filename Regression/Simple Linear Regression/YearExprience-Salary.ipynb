# Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

random_state = 49
```

# Preprocessing

```python
#Carga los datos y muestras las 10 primeras filas
datas = pd.read_csv("Salary_Data.csv")
datas.head()
```

```python
#Crea las variables dependientes e independientes
X = datas.iloc[:, :-1].values
y = datas.iloc[:, -1].values
```

```python
#Muestra en la gráfica los datos
plt.scatter(X, y, c=y)

plt.xlabel("Years Expreience")
plt.ylabel("Salary")
plt.show()
```

## Traing and test datasets

```python
#Se divide el dataset en datos de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

print("Dimensión de X_train:")
print(X_train.shape)
print("Dimensión de X_test:")
print(X_test.shape)
print("Dimensión de y_train:")
print(y_train.shape)
print("Dimensión de y_test:")
print(y_test.shape)
```

# Entrenamiento del modelo

```python
model = LinearRegression().fit(X_train, y_train)
```

# Predicciones

```python
y_pred = model.predict(X_test)
```

```python
# Plot del resultado
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, model.predict(X_train), color="blue", linewidth=2)

plt.title("Salary vs Experience (Training set)")
plt.xlabel("Years Expreience")
plt.ylabel("Salary")
plt.show()
```

```python
# Plot del resultado
plt.scatter(X_test, y_test, color="red")
plt.plot(X_test, y_pred, color="blue", linewidth=2)

plt.title("Salary vs Experience (Test set)")
plt.xlabel("Years Expreience")
plt.ylabel("Salary")
plt.show()
```

# Model Evaluation

```python
# The coefficients
print("Coefficients: \n", model.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
```

# Final results

```python
#Salario para 5 años de experiencia
salary = model.predict([[5],])

print("El salario para 5 años de exteriencia es de %.2f" % salary)


print("La ecuación del modelo es Salary = %.2f * YearExperience + %.2f" %(model.coef_, model.intercept_)) 
```
