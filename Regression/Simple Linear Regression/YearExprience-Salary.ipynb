# Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
random_state = 49
```

# Preprocessing

```python
#Carga los datos y muestras las 10 primeras filas
datas = pd.read_csv("Salary_Data.csv")
datas.head()
```

```python
#Crea las variables dependientes e independientes
X = datas.iloc[:, 0].array
y = datas.iloc[:, 1].array
```

```python
#Muestra en la gráfica los datos
plt.scatter(X, y, c=y)

plt.xlabel("Years Expreience")
plt.ylabel("Salary")
plt.show()
```

## Traing and test datasets

```python
#Se divide el dataset en datos de entrenamiento y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)

print("Dimensión de X_train:")
print(X_train.shape)
print("Dimensión de X_test:")
print(X_test.shape)
print("Dimensión de y_train:")
print(y_train.shape)
print("Dimensión de y_test:")
print(y_test.shape)
```

# Entrenamiento del modelo

```python
#Debido a que LinearRegression precisa que X_train tenga diemnsión 2, se redimensiona X_train e X_test
X_train_rd = np.reshape(X_train, (-1,1))
X_test_rd = np.reshape(X_test, (-1,1))
print(X_train.shape)
```

```python
model = LinearRegression().fit(X_train_rd, y_train)
```

# Predicciones

```python
y_pred = model.predict(X_test_rd)
```

```python
# Plot del resultado
plt.scatter(X_test_rd, y_test, color="red")
plt.plot(X_test_rd, y_pred, color="blue", linewidth=2)

plt.xlabel("Years Expreience")
plt.ylabel("Salary")
plt.show()
```
